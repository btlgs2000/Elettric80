```python
import optuna
from optuna import Trial
```

```python
def objective(trial):
    x = trial.suggest_float('x', -1, 1)
    y = trial.suggest_float('y', -1, 1)
    return x**2 + y**2
```

```python
study = optuna.create_study(storage='sqlite:///optuna.db', study_name='primo_studio', direction='minimize', load_if_exists=True)
```

```python
study.optimize(func=objective, n_trials=100)
```

```python
study.best_params
```

```python
study.best_trial
```
